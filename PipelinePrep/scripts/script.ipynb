```python
datas = pd.DataFrame(index = grouped_index).reset_index()
datas[(datas['SHIP_ID']=='T20220323003') & (datas['OP_INDEX']==20) & (datas['SECTION']==0)]
```

```python
for rows in grouped_index[45129:]:  
    
    try:
        ship_id = rows[0]
        op_index = rows[1]
        section = rows[2]

        df = data[(data['SHIP_ID']==ship_id) & (data['OP_INDEX']==op_index) & (data['SECTION']==section)]
        data_len = len(df)

        print(f'SHIP_ID : {ship_id} / OP_INDEX : {op_index} / SECTION : {section} -  데이터 선택 ({data_len})')

        if data_len>=100:
            print(f'SHIP_ID : {ship_id} / OP_INDEX : {op_index} / SECTION : {section} -  조건 통과')
            
            distribute_by_application(ship_id,op_index,section)
                                      
    except ZeroDivisionError:
        print(f"Skipping division by zero for {rows}")
        continue  # 오류가 발생하면 다음 반복으로 넘어감
        
    except ValueError:
        print(f"Skipping invalid data: {rows}")
        continue  # 오류가 발생하면 다음 반복으로 넘어감
        
    except KeyError:
        print(f"Skipping invalid data: {rows}")
        continue  # 오류가 발생하면 다음 반복으로 넘어감
        
    except IndexError:
        print(f"Skipping invalid data: {rows}")
        continue  # 오류가 발생하면 다음 반복으로 넘어감
```
